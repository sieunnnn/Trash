## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df.dropna() # 전체 데이터 중에서 NaN 을 포함하는 데이터를 삭제
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

- axis : index or columns
- how : any or all

```python
df.dropna(axis="index", how='any') # NaN 가 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') # NaN 이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis="columns", how='all') # 데이터 전체가 NaN 인 경우에만 column 삭제
```
